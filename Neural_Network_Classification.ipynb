```python
print(X_train.shape,y_train.shape)
```

```python
def one_hot_encode_object_array(arr):
  uniques, ids = np.unique(arr, return_inverse=True)
  return np_utils.to_categorical(ids, len(uniques))
```

```python
y_train_ohe = one_hot_encode_object_array(y_train)
y_test_ohe = one_hot_encode_object_array(y_test)
```

```python
model=Sequential()

model.add(Dense(5, input_shape=(4,), activation='relu', name='layer1'))

model.add(Dense(3, activation='sigmoid', name='output_layer'))

model.add(Dense(5,input_shape=(4,)))
model.add(Activation("relu",name="layer1"))
model.add(Dense(3))
model.add(Activation("sigmoid"))

model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
model.summary()
```

```python
model.fit(X_train,y_train_ohe,epochs=20,batch_size=1,verbose=1,validation_split=0.2)
```

```python
loss, accuracy = model.evaluate(X_test, y_test_ohe, verbose=1)
print("The accuracy of the Neural Network classifier on test data is = {:.2f}".format(accuracy))
```
